```python
def get_text_from_url(url):
    page = urllib.request.urlopen(url).read().decode('utf8','ignore')
    soup = BeautifulSoup(page, 'lxml')
    text = ' '.join(map(lambda p: p.text, soup.find_all('article')))
    return text.encode('ascii','replace').decode().replace("?", ' ')

text = get_text_from_url(article_url)
text
```

```python
from string import punctuation
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

sents = sent_tokenize(str(text))
sents
```

```python
word_sents = word_tokenize(text)
word_sents
```

```python
_stop_words = set(stopwords.words('english') + list(punctuation))
_stop_words
```

```python
word_sents = [word for word in word_sents if word not in _stop_words]
word_sents
```

```python
from nltk.probability import FreqDist
freq = FreqDist(word_sents)
freq
```

```python
from heapq import nlargest
```

```python
nlargest(10, freq, key=freq.get)
```

```python
from collections import defaultdict

ranking = defaultdict(int)

for i, sent in enumerate(sents):
    for w in word_tokenize(sent):
        if w in freq:
            ranking[i] += freq[w]
    
ranking
```

```python
sent_idx = nlargest(4, ranking, key=ranking.get)
sent_idx
```

```python
[sents[i] for i in sorted(sent_idx)]
```

```python
def summarize(text, n):
    sents = sent_tokenize(str(text))
    
    assert n <= len(sents)
    word_sents = word_tokenize(text.lower())
    _stop_words = set(stopwords.words('english') + list(punctuation))
    
    
    word_sents = [word for word in word_sents if word not in _stop_words]
    freq = FreqDist(word_sents)
    
    ranking = defaultdict(int)

    for i, sent in enumerate(sents):
        for w in word_tokenize(sent):
            if w in freq:
                ranking[i] += freq[w]
                
    sent_idx = nlargest(4, ranking, key=ranking.get)
    return [sents[i] for i in sorted(sent_idx)]
```

```python
summarize(text, 4)
```
